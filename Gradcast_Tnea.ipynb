```python
pred_College_Code = pd.read_csv("tnea2.csv")
pred_College_Code.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
pred_College_Code['COMMUNITY'] = le.fit_transform(pred_College_Code['COMMUNITY'])
pred_College_Code['COMMUNITY'].unique()
pred_College_Code['COLLEGE_CODE'] = le.fit_transform(pred_College_Code['COLLEGE_CODE'])
pred_College_Code['COLLEGE_CODE'].unique()
pred_College_Code['BRANCH_CODE'] = le.fit_transform(pred_College_Code['BRANCH_CODE'])
pred_College_Code['BRANCH_CODE'].unique()
pred_College_Code['ALL_CATEGORY'] = le.fit_transform(pred_College_Code['ALL_CATEGORY'])
pred_College_Code['ALL_CATEGORY'].unique()
```

```python
y = pred_College_Code[['COLLEGE_CODE', 'BRANCH_CODE', 'ALL_CATEGORY']]
```

```python
X = pred_College_Code[['AGGR_MARK', 'RANK', 'COMMUNITY']]
```

```python
# Spilliting the data into training and testing dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
# Importing library for model generation
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
# Predicting the model again in numerical terms

y_pred = model.predict(X_test)
y_pred
```

```python
# Checking the Mean Absolute error of whole model

from sklearn.metrics import mean_absolute_error
print("mean_absolute_error : ", mean_absolute_error(y_pred, y_test))
```

```python
# Accuracy for Institute short

from sklearn.metrics import classification_report 
y_pred1 = y_pred[:,0]
y_test1 = y_test.loc[:,"COLLEGE_CODE"]
print(classification_report(y_pred1, y_test1))
```

```python
mean_absolute_error(y_pred1, y_test1)
```

```python
from sklearn.metrics import accuracy_score
Accuracy1 = accuracy_score(y_pred1, y_test1)
Accuracy1
```

```python
# Accuracy for Program Name

from sklearn.metrics import classification_report 
y_pred2 = y_pred[:,1]
y_test2 = y_test.loc[:,"BRANCH_CODE"]
print(classification_report(y_pred2, y_test2))
```

```python
mean_absolute_error(y_pred2, y_test2)
```

```python
from sklearn.metrics import accuracy_score
Accuracy2 = accuracy_score(y_pred2, y_test2)
Accuracy2
```

```python
# Accuracy for Degree 

from sklearn.metrics import classification_report 
y_pred3 = y_pred[:,2]
y_test3 = y_test.loc[:,"ALL_CATEGORY"]
print(classification_report(y_pred3, y_test3))
```

```python
mean_absolute_error(y_pred3, y_test3)
```

```python
from sklearn.metrics import accuracy_score
Accuracy3 = accuracy_score(y_pred3, y_test3)
Accuracy3
```

```python
# total accuracy of the model

Mean_Accuracy = (Accuracy1 + Accuracy2 + Accuracy3) / 3
Mean_Accuracy
```
